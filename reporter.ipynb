```python
import pandas as pd
```

```python
i=pd.read_csv('pangolin_bs.csv')
i
```

```python
nf1=pd.read_csv('pangolin_IFO.csv')
nf2= pd.read_csv('pangolin_spall.csv')
```

```python
s=[str(barcode) for barcode in nf1['sample']]
```

```python
nf1['sample']=s
```

```python
nf=pd.concat([nf1,nf2]).reset_index()
```

Check on samples

```python
set(i.Taxon).difference(set(nf['sample']))
```

```python
len(i)-len(nf)
```

### Merging two tables

```python
out=i.merge(nf,right_on='sample', left_on='Taxon')[['Taxon','Lineage','lineage']]
```

```python
out.columns = ['sample','bs','nf']
```

```python
out[out.bs=='None']
```

```python
out[out.nf=='None']
```

#### Proportion of matching results

```python
len(out[out.bs == out.nf])/len(out)
```

#### Discrepancy

```python
divergent = out[out.bs != out.nf]
divergent
```

```python
counter = 0
for row in divergent.iterrows():
    if row[1].bs[:3] == row[1].nf[:3]:
        counter+=1
```

```python
print(len(divergent), 'samples have non-matching lineages:')
print('- Nf-core did not assign lineage to',len(divergent[divergent['nf']=='None']),'samples, of which', len(divergent[divergent['nf']=='None'][divergent['bs']=='A']) + len(divergent[divergent['nf']=='None'][divergent['bs']=='B']), 'have a generic lineage assigned in basespace')
print('-',counter, 'samples have partial identity')
```

### Merging with coverage info

```python
cov1=pd.read_csv('coverage_bs.tsv',sep='\t')
cov2=pd.read_csv('coverage_ifo.csv',sep='\t')
cov3=pd.read_csv('coverage_spall.tsv',sep='\t')
cov1
```

```python
cov=pd.concat([cov2,cov3])
```

```python
cov_diff = cov2.merge(cov,left_on='Sample',right_on='Sample')
cov_diff
```

```python
cov_diff['rbd'] = cov_diff.RBD_max_coverage_x - cov_diff.RBD_max_coverage_y + cov_diff.RBD_min_coverage_x - cov_diff.RBD_min_coverage_y + cov_diff.RBD_median_coverage_x - cov_diff.RBD_median_coverage_y
```

```python
cov_diff['spike'] = cov_diff.Spike_max_coverage_x - cov_diff.Spike_max_coverage_y + cov_diff.Spike_min_coverage_x - cov_diff.Spike_min_coverage_y + cov_diff.Spike_median_coverage_x - cov_diff.Spike_median_coverage_y
```

```python
cov_diff.rbd.mean()
```

```python
cov_diff.spike.mean()
```
