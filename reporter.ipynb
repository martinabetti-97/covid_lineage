### Merging two tables

```python
out=i.merge(nf,right_on='sample', left_on='taxon')[['taxon_x','lineage_x','lineage_y']]
```

```python
out
```

```python
out.columns = ['sample','bs','nf']
```

```python
out['bs']=out['bs'].str.replace('AY.12','B.1.617.2')
out['bs']=out['bs'].str.replace('AY.9','B.1.617.2')
out['bs']=out['bs'].str.replace('AY.4','B.1.617.2')
```

```python
out[out.bs=='None']
```

```python
out[out.nf=='None']
```

#### Proportion of matching results

```python
len(out[out.bs == out.nf])/len(out)
```

#### Discrepancy

```python
divergent = out[out.bs != out.nf]
divergent
```

```python
counter = 0
for row in divergent.iterrows():
    if row[1].bs[:3] == row[1].nf[:3]:
        counter+=1
```

```python
print(len(divergent), 'samples have non-matching lineages:')
print('- Nf-core did not assign lineage to',len(divergent[divergent['nf']=='None']),'samples, of which', len(divergent[divergent['nf']=='None'][divergent['bs']=='A']) + len(divergent[divergent['nf']=='None'][divergent['bs']=='B']), 'have a generic lineage assigned in basespace')
print('-',counter, 'samples have partial identity')
```

### Merging with coverage info

```python
cov1=pd.read_csv('coverage_0309_bs.tsv',sep='\t')
cov2=pd.read_csv('coverage_0309_nf.tsv',sep='\t')
```

```python
cov1['Sample']=cov1['Sample'].str.replace('-','_')
```

```python
cov2['Sample']=cov2['Sample'].str.replace('-','_')
```

```python
cov_diff = cov2.merge(cov1,left_on='Sample',right_on='Sample')
cov_diff
```

```python
cov_diff['rbd'] = cov_diff.RBD_median_coverage_x - cov_diff.RBD_median_coverage_y
```

```python
cov_diff['spike'] = cov_diff.Spike_median_coverage_x - cov_diff.Spike_median_coverage_y
```

```python
cov_diff.rbd.mean()
```

```python
cov_diff.spike.mean()
```
